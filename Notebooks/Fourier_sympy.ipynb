```python
from sympy import fourier_series, pi, plot, Piecewise
from sympy.abc import x
```

```python
f=Piecewise(
    (0, x < -2),
    (x+2,(x >= -2) & (x < 0)),
    (0,(x<=2)))
```

```python
plot(f, (x,-2,2),line_color ='red')
```

```python
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

```python
from sympy import exp, pi, cos, integrate, fourier_series
from sympy.abc import x
from sympy.plotting import plot

f = exp(2*x)

a_0 = 2 * integrate(f, (x, -1, 1))

n_terms = 7
a_n = []
for n in range(1, n_terms + 1):
    coef = 2 * integrate(f * cos(n * pi * x), (x, -1, 1))
    a_n.append(coef)

fourier_series_cos = a_0 / 2
for n, coef in enumerate(a_n, 1):
    fourier_series_cos += coef * cos(n * pi * x)

p = plot(f, fourier_series_cos, (x, -1, 1), show=False, legend=True)

p[0].label = 'e^(2x)'

p[1].label = 'Serie en cosenos'

p.show()
```
