```python
import numpy as np
import matplotlib.pyplot as plt

c = 1  # velocidad de propagacin
t = 0.2  # tiempo
x = np.linspace(0, 1, 1000)

# Funcin original f(x') con x' = x
def f(xp):
    return np.exp(-((xp - 0.5) ** 2) / 0.01)

# Onda que se desplaza a la derecha: x' = x - ct
f_derecha = f(x - c * t)

# Onda que se desplaza a la izquierda: x' = x + ct
f_izquierda = f(x + c * t)

# Superposicin
f_superpuesta = f_derecha + f_izquierda

# Grficas a), b) y c)
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, f_derecha, label="Onda a la derecha")
plt.title("Desplazamiento a la derecha")
plt.xlabel("x")
plt.ylabel("f(x - ct)")
plt.grid()
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, f_izquierda, label="Onda a la izquierda", color='orange')
plt.title("Desplazamiento a la izquierda")
plt.xlabel("x")
plt.ylabel("f(x + ct)")
plt.grid()
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, f_superpuesta, label="Superposicin", color='green')
plt.title("Superposicin de ondas")
plt.xlabel("x")
plt.ylabel("f(x - ct) + f(x + ct)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
