```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.piecewise(x,
                        [(-2 <= x) & (x < 0), (0 <= x) & (x < 2)], #aqu definimos los la funcin a trozos
                        [0, lambda x: 2 - x, 0])

L=2
N=10

a0=(1/L)*np.trapz([f(x) for x in np.linspace(-L,L, 1000)], np.linspace(-L,L,1000))  #se hacen los clculos para la integral que se usa para calcular el a0
an=[]
bn=[]
for n in range(1,N+1):
  an.append((1/L)*np.trapz([f(x)*np.cos(n*np.pi*x/L) for x in np.linspace(-L,L,1000)],np.linspace(-L,L,1000)))
  bn.append((1/L)*np.trapz([f(x)*np.sin(n*np.pi*x/L) for x in np.linspace(-L,L,1000)],np.linspace(-L,L,1000)))  #Tanto en el paso anterior como en este se definen las integrales respectivas de an y bn



x=np.linspace(-2.5,2.5,500)
y=a0/2
for n in range(1,N+1):
  y += an[n-1]*np.cos(n*np.pi*x/L)+bn[n-1]*np.sin(n*np.pi*x/L) #Aqu ya est toda la serie de Fourier definida

#En todo esto son los parmetros para la grfica. Ttulos, etiquetas y funciones para x y y.
#Finalmente se ensea la grfica
plt.figure(figsize=(10,6))
plt.plot(x,[f(i) for i in x], label='Funcin original')
plt.plot(x,y,label=f'Serie de Fourier(N={N})')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Serie de Fourier para una funcin a trozos')
plt.legend()
plt.grid(True)
plt.show()
```
